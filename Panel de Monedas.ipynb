```python
from pycoingecko import CoinGeckoAPI
cg = CoinGeckoAPI()
```

```python
import pandas as pd
markets_data = cg.get_coins_markets(vs_currency="usd", price_change_percentage="200d")
dframe = pd.DataFrame(markets_data)
dframe.to_csv("data_markets.csv")
```

```python
dframe = pd.read_csv('data_markets.csv')
dframe.head(1)
```

```python
columnas_seleccionadas = ['id', 'symbol', "current_price", "market_cap", "market_cap_rank","total_volume", "ath", "atl", "price_change_percentage_200d_in_currency"]
dframe = dframe[columnas_seleccionadas]
dframe.head()

criptos = ['btc', 'eth', 'dot', 'usdc', 'dai', 'mana', 'usdt', 'xmr', 'ada', 'sol']
```

```python
dframe = dframe[dframe['symbol'].isin(criptos)]
dframe.head(11)
```

```python
dframe["total_volume"] = pd.to_numeric(dframe["total_volume"])
def format_number(num):
    return '{:,.0f}'.format(num)
dframe["total_volume"] = dframe["total_volume"].apply(format_number)

dframe.head(1)
```

```python
dframe.rename(columns={"ath": "MÃ¡ximo HistÃ³rico"}, inplace=True)
dframe.rename(columns={"atl": "MÃ­nimo HistÃ³rico"}, inplace=True)
dframe.rename(columns={"price_change_percentage_200d_in_currency": "Media 200D"}, inplace=True)
dframe.head(11)
```

```python
dframe["MÃ­nimo HistÃ³rico"] = dframe["MÃ­nimo HistÃ³rico"].round(2)
dframe["MÃ¡ximo HistÃ³rico"] = dframe["MÃ¡ximo HistÃ³rico"].round(2)
dframe["Media 200D"] = dframe["Media 200D"].round(2)
```
